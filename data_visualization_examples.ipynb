```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random points for X and Y (0-100)
X_data = np.random.random(1000) * 100
Y_data = np.random.random(1000) * 100

# Plot a scatter plot with red star markers and semi-transparency
plt.scatter(X_data, Y_data, c="red", marker="*", alpha=0.5)

# Show the plot
plt.show()
```

```python
# Create a list of years from 2006 to 2021
years = [2006 + x for x in range(16)]
print(years)
# List of weights corresponding to each year
weights = [80, 83, 85, 86, 82, 81, 79, 83, 80, 82, 82, 83, 81, 80, 79,78]
# Plot a line graph:
# - x-axis: years
# - y-axis: weights
plt.plot(years, weights, c='g', linewidth=3)
# Display the plot
plt.show()
```

```python
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 2, 45]

# Create a bar chart:
# - x-axis: programming languages
# - y-axis: number of votes
# - bar width = 0.4
# - edge color of bars = green
plt.bar(x, y, width=0.4, edgecolor="green")

# Add title and labels
plt.title("Number of votes for languages")
plt.xlabel("Language")
plt.ylabel("number of votes")
# Display the plot
plt.show()
```

```python
# Generate 1000 random ages with a normal distribution
# - mean = 20
# - standard deviation = 1.5
ages = np.random.normal(20, 1.5, 1000)

# Plot a histogram with custom bins:
# - bin1: from minimum age to 18
# - bin2: from 18 to 21
# - bin3: from 21 to maximum age
plt.hist(ages,bins=[ages.min(), 18, 21, ages.max()])

# Add axis labels and title
plt.xlabel("Age")
plt.ylabel("Nombre de personnes")
plt.title("Histogramme avec intervalles personnalises")

plt.show()
```

```python
langs = ["Python", "C++", "Java", "C#", "Go"]
votes = [50, 24, 14, 6, 17]
# Define which slice to "explode" (offset from center)
# Here, the 4th slice (C#) will be slightly separated
explodes =[0, 0, 0, 0.1, 0]

# Create a pie chart:
# - votes = sizes of each slice
# - labels = names of languages
# - explode = offset specific slices
# - autopct = show percentage with 2 decimal places
# - pctdistance = distance of percentage labels from the center
# - startangle = start rotation of the pie (180 degrees)
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%", pctdistance=1.3, startangle=180)

plt.show()
```

```python
# Generate 300 random heights with a normal distribution
# - mean = 170 cm
# - standard deviation = 8 cm
heights = np.random.normal(170, 8, 300)

# Create a boxplot to visualize the distribution
# - shows median, quartiles, whiskers, and potential outliers
plt.boxplot(heights)
plt.show()
```

```python
first = np.linspace(0, 10, 25)
second = np.linspace(10, 200, 25)
third = np.linspace(200, 210, 25)
fourth = np.linspace(210, 230, 25)

data =np.concatenate((first, second, third, fourth))

plt.boxplot(data)
plt.show()
```

```python
# List of years
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

# John's income for each year
income = [55, 56, 62, 61, 72, 72, 73, 75]

# Create a list of ticks for the y-axis (from 50 to 80, step 2)
income_tickets = list(range(50, 81, 2))

# Plot a line graph of income over years
plt.plot(years, income)

# Set the title and axis labels with font size
plt.title("Income of John (in USD)", fontsize=20)
plt.xlabel("Year")
plt.ylabel("Income in USD")

# Customize y-axis ticks with formatted labels
plt.yticks(income_tickets, [f"${x}k USD" for x in income_tickets])

# Display the plot
plt.show()
```

```python
# Stock prices for three companies over 7 days
stock_a = [100, 102, 99, 101, 101, 100, 102]
stock_b = [90, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]

# Plot the stock prices as line graphs
# - label specifies the company for the legend
plt.plot(stock_a, label="Company 1")
plt.plot(stock_b, label="Company 2")
plt.plot(stock_c, label="Company 3")

# Display a legend in the lower right corner
plt.legend(loc="lower right")

# Display the plot
plt.show()
```

```python
# Number of votes for each person
votes = [10, 2, 5, 16, 22]

# Names of the people
people = ["A", "B", "C", "D", "E"]

# Create a pie chart
# - slices sizes = votes
# - labels = people's names
plt.pie(votes, labels=people)

# Add a legend using the same labels
plt.legend(labels=people)

# Display the plot
plt.show()
```

```python
# Generate random data for scatter plot
x1, y1 = np.random.random(100), np.random.random(100)

# Generate data for line plot
x2, y2 = np.arange(100), np.random.random(100)

# Create the first figure (Figure 1) for scatter plot
plt.figure(1)
plt.scatter(x1, y1)  # Scatter plot with random points

# Create the second figure (Figure 2) for line plot
plt.figure(2)
plt.plot(x2, y2)  # Line plot with sequential x values and random y values

# Display both figures
plt.show()
```

```python
# Generate x values from 0 to 99
x = np.arange(100)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Top-left subplot: plot sin(x)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sin(x)")

# Top-right subplot: plot cos(x)
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cos(x)")

# Bottom-left subplot: plot random values
axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

# Bottom-right subplot: plot log(x)
# Note: log(0) is -inf, so it may give a warning
axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("log(x)")
axs[1, 1].set_xlabel("Test")

# Add an overall title for the figure
fig.suptitle("Four Plots")

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the figure as an image
plt.savefig("fourplots.png")

# Display all subplots
plt.show()
```
