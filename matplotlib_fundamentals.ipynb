```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Basic Graph**

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize the figure (width=5, height=3) with 100 DPI
plt.figure(figsize=(5,3), dpi=100)

# First line: y = 2x, with style customization
plt.plot(x, y, label='2x', color='red', linestyle='--', linewidth=3,
         marker='o', markersize=5, markeredgecolor='green')

# Second line
# start = 0 → starts at 0
# stop = 4 → stops before 4
# step = 0.5 → increments by 0.5 each time
x2 = np.arange(0,4,0.5)
plt.plot(x2[:5], x2[:5]**2, 'y', label='x**2')

# Titles and labels
plt.title("Our First Graph", fontdict={'fontsize': 15})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Set ticks manually
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

# Show legend
plt.legend()

# Save the figure as a PNG with 300 DPI
plt.savefig('mygraph.png', dpi=300)

# Display the plot
plt.show()
```

**Bar Chart**

```python
labels = ['A', 'B', 'C']
values =[1,4,2]

bars = plt.bar(labels, values)

patterns= ['/', 'o', '*']

for bar in bars:
    bar.set_hatch(patterns.pop())

# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('o')

plt.figure(figsize=(6,4))
plt.show()
```
